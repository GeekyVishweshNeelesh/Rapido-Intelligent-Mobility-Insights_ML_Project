```python
"""
Cell 5: Convert Data Types for Feature Engineering
"""

print("🔧 CONVERTING DATA TYPES")
print("="*70)

# Convert datetime columns in bookings
datetime_cols = [col for col in df_bookings.columns
                 if any(keyword in col.lower() for keyword in ['date', 'time'])]

print("\n📅 Converting datetime columns:")
for col in datetime_cols:
    if col in df_bookings.columns:
        try:
            df_bookings[col] = pd.to_datetime(df_bookings[col], errors='coerce')
            print(f"   ✅ {col}")
        except:
            print(f"   ⚠️ {col} - conversion failed")

# Identify key columns
print("\n" + "="*70)
print("📋 IDENTIFYING KEY COLUMNS:")
print("="*70)

# Status column
status_col = None
for col in df_bookings.columns:
    if 'status' in col.lower():
        status_col = col
        print(f"   Status column: {status_col}")
        break

# Fare column
fare_col = None
for col in df_bookings.columns:
    if any(keyword in col.lower() for keyword in ['value', 'fare', 'price', 'amount']):
        fare_col = col
        print(f"   Fare column: {fare_col}")
        break

# Distance column
distance_col = None
for col in df_bookings.columns:
    if 'distance' in col.lower():
        distance_col = col
        print(f"   Distance column: {distance_col}")
        break

print("\n✅ Data types configured!")
print("="*70)
```

```python
"""
Cell 7: Merge Customer Information
"""

print("👥 MERGING CUSTOMER DATA")
print("="*70)

if 'customer_id' in df_bookings.columns and 'customer_id' in df_customers.columns:

    print(f"\n📊 Before merge:")
    print(f"   Bookings: {len(df_bookings):,} rows")
    print(f"   Unique customers in bookings: {df_bookings['customer_id'].nunique():,}")
    print(f"   Unique customers in master: {df_customers['customer_id'].nunique():,}")

    # Rename customer columns to avoid conflicts
    customer_cols_to_rename = {col: f'cust_{col}' for col in df_customers.columns
                               if col != 'customer_id'}
    df_customers_renamed = df_customers.rename(columns=customer_cols_to_rename)

    print(f"\n📋 Customer columns to merge: {list(customer_cols_to_rename.values())}")

    # Perform merge
    df_master = df_bookings.merge(
        df_customers_renamed,
        on='customer_id',
        how='left',
        suffixes=('', '_cust_dup')
    )

    print(f"\n✅ After merge:")
    print(f"   Rows: {len(df_master):,} (should match bookings)")
    print(f"   Columns: {len(df_master.columns)} (was {len(df_bookings.columns)})")
    print(f"   New columns added: {len(df_master.columns) - len(df_bookings.columns)}")

    # Verify merge quality
    if len(df_master) == len(df_bookings):
        print(f"\n   ✅ Merge successful - no row loss!")
    else:
        print(f"\n   ⚠️ Warning: Row count changed!")

    # Check for nulls in merged columns
    new_cols = [col for col in df_master.columns if col.startswith('cust_')]
    if new_cols:
        null_count = df_master[new_cols].isnull().sum().sum()
        if null_count > 0:
            print(f"   ⚠️ {null_count:,} null values in customer data (customers not in master table)")
        else:
            print(f"   ✅ No nulls in customer data")
else:
    print("\n⚠️ Customer ID column not found")
    df_master = df_bookings.copy()

print("\n" + "="*70)
```

```python
"""
Cell 8: Merge Driver Information
"""

print("🚕 MERGING DRIVER DATA")
print("="*70)

if 'driver_id' in df_master.columns and 'driver_id' in df_drivers.columns:

    print(f"\n📊 Before merge:")
    print(f"   Current rows: {len(df_master):,}")
    print(f"   Unique drivers in bookings: {df_master['driver_id'].nunique():,}")
    print(f"   Unique drivers in master: {df_drivers['driver_id'].nunique():,}")

    # Rename driver columns
    driver_cols_to_rename = {col: f'driver_{col}' for col in df_drivers.columns
                             if col != 'driver_id'}
    df_drivers_renamed = df_drivers.rename(columns=driver_cols_to_rename)

    print(f"\n📋 Driver columns to merge: {list(driver_cols_to_rename.values())}")

    # Perform merge
    df_master_before = len(df_master)
    df_master = df_master.merge(
        df_drivers_renamed,
        on='driver_id',
        how='left',
        suffixes=('', '_drv_dup')
    )

    print(f"\n✅ After merge:")
    print(f"   Rows: {len(df_master):,}")
    print(f"   Columns: {len(df_master.columns)}")

    if len(df_master) == df_master_before:
        print(f"   ✅ Merge successful - no row loss!")
    else:
        print(f"   ⚠️ Warning: Row count changed!")

    # Check for nulls
    new_cols = [col for col in df_master.columns if col.startswith('driver_')]
    if new_cols:
        null_count = df_master[new_cols].isnull().sum().sum()
        if null_count > 0:
            print(f"   ⚠️ {null_count:,} null values in driver data")
        else:
            print(f"   ✅ No nulls in driver data")
else:
    print("\n⚠️ Driver ID column not found")

print("\n" + "="*70)
```

```python
"""
Cell 11: Extract Basic Time Features from Datetime Columns
"""

print("⏰ CREATING TIME-BASED FEATURES")
print("="*70)

# Find datetime columns
datetime_cols = [col for col in df_master.columns
                 if pd.api.types.is_datetime64_any_dtype(df_master[col])]

print(f"\n📅 Datetime columns found: {datetime_cols}")

# Use booking_time or booking_date as primary time column
time_col = None
for col in ['booking_time', 'booking_date', 'booking_datetime']:
    if col in df_master.columns:
        time_col = col
        break

if not time_col and len(datetime_cols) > 0:
    time_col = datetime_cols[0]

if time_col:
    print(f"\n🕐 Using primary time column: {time_col}")

    # Ensure it's datetime
    df_master[time_col] = pd.to_datetime(df_master[time_col], errors='coerce')

    # Extract features
    print(f"\n📊 Creating time features...")

    # Hour
    if 'hour' not in df_master.columns:
        df_master['hour'] = df_master[time_col].dt.hour
        print(f"   ✅ hour (0-23)")

    # Day of week
    if 'day_of_week' not in df_master.columns:
        df_master['day_of_week'] = df_master[time_col].dt.dayofweek
        print(f"   ✅ day_of_week (0=Mon, 6=Sun)")

    # Month
    if 'month' not in df_master.columns:
        df_master['month'] = df_master[time_col].dt.month
        print(f"   ✅ month (1-12)")

    # Day of month
    if 'day_of_month' not in df_master.columns:
        df_master['day_of_month'] = df_master[time_col].dt.day
        print(f"   ✅ day_of_month (1-31)")

    # Quarter
    if 'quarter' not in df_master.columns:
        df_master['quarter'] = df_master[time_col].dt.quarter
        print(f"   ✅ quarter (1-4)")

    # Year
    if 'year' not in df_master.columns:
        df_master['year'] = df_master[time_col].dt.year
        print(f"   ✅ year")

    # Week of year
    if 'week_of_year' not in df_master.columns:
        df_master['week_of_year'] = df_master[time_col].dt.isocalendar().week
        print(f"   ✅ week_of_year (1-53)")

    print(f"\n✅ Created {7} basic time features")

else:
    print("\n⚠️ No datetime column found - skipping time features")

print("\n" + "="*70)
print(f"Current dataset shape: {df_master.shape}")
print("="*70)
```

```python
"""
Cell 12: Create Advanced Time Features (Rush Hours, Weekends, etc.)
"""

print("⏰ CREATING ADVANCED TIME FEATURES")
print("="*70)

features_created = []

# First, ensure day_of_week and hour are numeric
if 'day_of_week' in df_master.columns:
    # Convert to numeric if it's not already
    df_master['day_of_week'] = pd.to_numeric(df_master['day_of_week'], errors='coerce')
    print(f"✅ day_of_week converted to numeric")

if 'hour' in df_master.columns:
    # Convert to numeric if it's not already
    df_master['hour'] = pd.to_numeric(df_master['hour'], errors='coerce')
    print(f"✅ hour converted to numeric")

print()

# 1. Is Weekend
if 'day_of_week' in df_master.columns:
    df_master['is_weekend'] = (df_master['day_of_week'] >= 5).astype(int)
    features_created.append('is_weekend')
    weekend_pct = (df_master['is_weekend'].sum() / len(df_master)) * 100
    print(f"✅ is_weekend (1=Sat/Sun, 0=Weekday) - {weekend_pct:.1f}% are weekends")

# 2. Is Rush Hour (Morning: 7-10, Evening: 17-20)
if 'hour' in df_master.columns:
    df_master['is_rush_hour'] = (
        ((df_master['hour'] >= 7) & (df_master['hour'] <= 10)) |
        ((df_master['hour'] >= 17) & (df_master['hour'] <= 20))
    ).astype(int)
    features_created.append('is_rush_hour')
    rush_pct = (df_master['is_rush_hour'].sum() / len(df_master)) * 100
    print(f"✅ is_rush_hour (morning 7-10 or evening 17-20) - {rush_pct:.1f}% are rush hour")

    # Rush hour type
    df_master['rush_hour_type'] = 'None'
    df_master.loc[(df_master['hour'] >= 7) & (df_master['hour'] <= 10), 'rush_hour_type'] = 'Morning'
    df_master.loc[(df_master['hour'] >= 17) & (df_master['hour'] <= 20), 'rush_hour_type'] = 'Evening'
    features_created.append('rush_hour_type')
    print(f"✅ rush_hour_type (Morning/Evening/None)")

# 3. Time of Day
if 'hour' in df_master.columns:
    df_master['time_of_day'] = 'Night'
    df_master.loc[(df_master['hour'] >= 6) & (df_master['hour'] < 12), 'time_of_day'] = 'Morning'
    df_master.loc[(df_master['hour'] >= 12) & (df_master['hour'] < 17), 'time_of_day'] = 'Afternoon'
    df_master.loc[(df_master['hour'] >= 17) & (df_master['hour'] < 21), 'time_of_day'] = 'Evening'
    features_created.append('time_of_day')
    print(f"✅ time_of_day (Morning/Afternoon/Evening/Night)")
    print(f"   Distribution:")
    print(df_master['time_of_day'].value_counts())

# 4. Is Late Night (22:00 - 5:00)
if 'hour' in df_master.columns:
    df_master['is_late_night'] = (
        (df_master['hour'] >= 22) | (df_master['hour'] <= 5)
    ).astype(int)
    features_created.append('is_late_night')
    late_night_pct = (df_master['is_late_night'].sum() / len(df_master)) * 100
    print(f"\n✅ is_late_night (22:00-5:00) - {late_night_pct:.1f}% are late night")

# 5. Is Business Hours (9-17 on weekdays)
if 'hour' in df_master.columns and 'day_of_week' in df_master.columns:
    df_master['is_business_hours'] = (
        (df_master['day_of_week'] < 5) &
        (df_master['hour'] >= 9) &
        (df_master['hour'] <= 17)
    ).astype(int)
    features_created.append('is_business_hours')
    biz_pct = (df_master['is_business_hours'].sum() / len(df_master)) * 100
    print(f"✅ is_business_hours (9-17 on weekdays) - {biz_pct:.1f}%")

# 6. Is Month End (last 3 days of month)
if 'day_of_month' in df_master.columns:
    df_master['day_of_month'] = pd.to_numeric(df_master['day_of_month'], errors='coerce')
    df_master['is_month_end'] = (df_master['day_of_month'] >= 28).astype(int)
    features_created.append('is_month_end')
    month_end_pct = (df_master['is_month_end'].sum() / len(df_master)) * 100
    print(f"✅ is_month_end (last 3 days) - {month_end_pct:.1f}%")

# 7. Is Month Start (first 3 days of month)
if 'day_of_month' in df_master.columns:
    df_master['is_month_start'] = (df_master['day_of_month'] <= 3).astype(int)
    features_created.append('is_month_start')
    month_start_pct = (df_master['is_month_start'].sum() / len(df_master)) * 100
    print(f"✅ is_month_start (first 3 days) - {month_start_pct:.1f}%")

# 8. Day Name (categorical)
if 'day_of_week' in df_master.columns:
    day_mapping = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday',
                   4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
    df_master['day_name'] = df_master['day_of_week'].map(day_mapping)
    features_created.append('day_name')
    print(f"✅ day_name (Monday-Sunday)")

# 9. Hour Category
if 'hour' in df_master.columns:
    df_master['hour_category'] = pd.cut(
        df_master['hour'],
        bins=[0, 6, 12, 18, 24],
        labels=['Early Morning', 'Morning', 'Afternoon-Evening', 'Night'],
        include_lowest=True
    )
    features_created.append('hour_category')
    print(f"✅ hour_category")

print(f"\n✅ Created {len(features_created)} advanced time features")
print(f"\n📊 Current shape: {df_master.shape}")
print("="*70)
```

```python
"""
Cell 13: Create Fare-Based Features
"""

print("💰 CREATING FARE-BASED FEATURES")
print("="*70)

# Identify fare and distance columns
fare_col = None
distance_col = None

for col in df_master.columns:
    if fare_col is None and any(kw in col.lower() for kw in ['fare', 'value', 'price', 'amount']):
        # Exclude columns that are clearly not fare (like customer/driver amounts)
        if not any(x in col.lower() for x in ['customer', 'driver', 'cust_', 'driver_']):
            fare_col = col
    if distance_col is None and 'distance' in col.lower():
        distance_col = col

print(f"\n📋 Using columns:")
print(f"   Fare: {fare_col}")
print(f"   Distance: {distance_col}")

# Convert fare and distance to numeric
if fare_col:
    df_master[fare_col] = pd.to_numeric(df_master[fare_col], errors='coerce')
    print(f"   ✅ {fare_col} converted to numeric")

if distance_col:
    df_master[distance_col] = pd.to_numeric(df_master[distance_col], errors='coerce')
    print(f"   ✅ {distance_col} converted to numeric")

features_created = []

# 1. Fare per KM
if fare_col and distance_col:
    df_master['fare_per_km'] = df_master[fare_col] / df_master[distance_col].replace(0, np.nan)
    df_master['fare_per_km'] = df_master['fare_per_km'].replace([np.inf, -np.inf], np.nan)
    features_created.append('fare_per_km')
    print(f"\n✅ fare_per_km")
    print(f"   Mean: ₹{df_master['fare_per_km'].mean():.2f}/km")
    print(f"   Median: ₹{df_master['fare_per_km'].median():.2f}/km")

# 2. Fare per Minute (if time available)
if fare_col and 'actual_ride_time_min' in df_master.columns:
    # Convert to numeric (handles both numeric and datetime)
    if pd.api.types.is_datetime64_any_dtype(df_master['actual_ride_time_min']):
        print(f"\n⚠️ actual_ride_time_min is datetime, skipping fare_per_min")
    else:
        df_master['actual_ride_time_min'] = pd.to_numeric(df_master['actual_ride_time_min'], errors='coerce')
        df_master['fare_per_min'] = df_master[fare_col] / df_master['actual_ride_time_min'].replace(0, np.nan)
        df_master['fare_per_min'] = df_master['fare_per_min'].replace([np.inf, -np.inf], np.nan)
        features_created.append('fare_per_min')
        print(f"✅ fare_per_min")
        print(f"   Mean: ₹{df_master['fare_per_min'].mean():.2f}/min")

# 3. Fare Category
if fare_col:
    # Calculate quartiles for better binning
    fare_25 = df_master[fare_col].quantile(0.25)
    fare_50 = df_master[fare_col].quantile(0.50)
    fare_75 = df_master[fare_col].quantile(0.75)

    df_master['fare_category'] = pd.cut(
        df_master[fare_col],
        bins=[0, fare_25, fare_50, fare_75, float('inf')],
        labels=['Low', 'Medium', 'High', 'Very High'],
        include_lowest=True
    )
    features_created.append('fare_category')
    print(f"\n✅ fare_category (based on quartiles)")
    print(f"   Distribution:")
    print(df_master['fare_category'].value_counts().sort_index())

# 4. Is High Fare (top 25%)
if fare_col:
    fare_75th = df_master[fare_col].quantile(0.75)
    df_master['is_high_fare'] = (df_master[fare_col] >= fare_75th).astype(int)
    features_created.append('is_high_fare')
    high_fare_pct = (df_master['is_high_fare'].sum() / len(df_master)) * 100
    print(f"\n✅ is_high_fare (>= ₹{fare_75th:.2f}) - {high_fare_pct:.1f}%")

# 5. Is Low Fare (bottom 25%)
if fare_col:
    fare_25th = df_master[fare_col].quantile(0.25)
    df_master['is_low_fare'] = (df_master[fare_col] <= fare_25th).astype(int)
    features_created.append('is_low_fare')
    low_fare_pct = (df_master['is_low_fare'].sum() / len(df_master)) * 100
    print(f"✅ is_low_fare (<= ₹{fare_25th:.2f}) - {low_fare_pct:.1f}%")

# 6. Surge Amount (if surge multiplier exists)
if 'surge_multiplier' in df_master.columns and 'base_fare' in df_master.columns:
    df_master['surge_multiplier'] = pd.to_numeric(df_master['surge_multiplier'], errors='coerce')
    df_master['base_fare'] = pd.to_numeric(df_master['base_fare'], errors='coerce')

    df_master['surge_amount'] = (df_master['surge_multiplier'] - 1) * df_master['base_fare']
    df_master['surge_amount'] = df_master['surge_amount'].clip(lower=0)  # No negative surge
    features_created.append('surge_amount')
    print(f"\n✅ surge_amount (extra due to surge)")
    print(f"   Mean surge amount: ₹{df_master['surge_amount'].mean():.2f}")

# 7. Has Surge
if 'surge_multiplier' in df_master.columns:
    df_master['surge_multiplier'] = pd.to_numeric(df_master['surge_multiplier'], errors='coerce')
    df_master['has_surge'] = (df_master['surge_multiplier'] > 1.0).astype(int)
    features_created.append('has_surge')
    surge_pct = (df_master['has_surge'].sum() / len(df_master)) * 100
    print(f"✅ has_surge - {surge_pct:.1f}% of rides")

# 8. Surge Category
if 'surge_multiplier' in df_master.columns:
    df_master['surge_category'] = pd.cut(
        df_master['surge_multiplier'],
        bins=[0, 1.0, 1.5, 2.0, float('inf')],
        labels=['No Surge', 'Low Surge', 'Medium Surge', 'High Surge'],
        include_lowest=True
    )
    features_created.append('surge_category')
    print(f"✅ surge_category")
    print(f"   Distribution:")
    print(df_master['surge_category'].value_counts().sort_index())

# 9. Fare Deviation (how much it deviates from mean)
if fare_col:
    fare_mean = df_master[fare_col].mean()
    fare_std = df_master[fare_col].std()
    df_master['fare_deviation'] = (df_master[fare_col] - fare_mean) / fare_std
    features_created.append('fare_deviation')
    print(f"\n✅ fare_deviation (z-score)")

# 10. Is Premium Fare (combination of high fare and surge)
if 'is_high_fare' in df_master.columns and 'has_surge' in df_master.columns:
    df_master['is_premium_fare'] = (
        (df_master['is_high_fare'] == 1) & (df_master['has_surge'] == 1)
    ).astype(int)
    features_created.append('is_premium_fare')
    premium_pct = (df_master['is_premium_fare'].sum() / len(df_master)) * 100
    print(f"✅ is_premium_fare (high fare + surge) - {premium_pct:.1f}%")

print(f"\n✅ Created {len(features_created)} fare-based features")
print(f"📊 Current shape: {df_master.shape}")
print("="*70)
```

```python
"""
Cell 14: Create Distance-Based Features
"""

print("📏 CREATING DISTANCE-BASED FEATURES")
print("="*70)

if distance_col:
    print(f"\nUsing distance column: {distance_col}\n")

    # Ensure numeric
    df_master[distance_col] = pd.to_numeric(df_master[distance_col], errors='coerce')

    features_created = []

    # Calculate statistics
    dist_25 = df_master[distance_col].quantile(0.25)
    dist_50 = df_master[distance_col].quantile(0.50)
    dist_75 = df_master[distance_col].quantile(0.75)

    # 1. Distance Category
    df_master['distance_category'] = pd.cut(
        df_master[distance_col],
        bins=[0, dist_25, dist_50, dist_75, float('inf')],
        labels=['Short', 'Medium', 'Long', 'Very Long'],
        include_lowest=True
    )
    features_created.append('distance_category')
    print(f"✅ distance_category (based on quartiles)")
    print(f"   Distribution:")
    print(df_master['distance_category'].value_counts().sort_index())

    # 2. Is Short Ride (< 25th percentile)
    df_master['is_short_ride'] = (df_master[distance_col] < dist_25).astype(int)
    features_created.append('is_short_ride')
    short_pct = (df_master['is_short_ride'].sum() / len(df_master)) * 100
    print(f"\n✅ is_short_ride (< {dist_25:.1f} km) - {short_pct:.1f}%")

    # 3. Is Long Ride (> 75th percentile)
    df_master['is_long_ride'] = (df_master[distance_col] > dist_75).astype(int)
    features_created.append('is_long_ride')
    long_pct = (df_master['is_long_ride'].sum() / len(df_master)) * 100
    print(f"✅ is_long_ride (> {dist_75:.1f} km) - {long_pct:.1f}%")

    # 4. Distance Percentile
    df_master['distance_percentile'] = pd.qcut(
        df_master[distance_col],
        q=10,
        labels=False,
        duplicates='drop'
    )
    features_created.append('distance_percentile')
    print(f"✅ distance_percentile (0-9 deciles)")

    # 5. Speed (if time available)
    if 'actual_ride_time_min' in df_master.columns:
        # Check if it's not datetime
        if not pd.api.types.is_datetime64_any_dtype(df_master['actual_ride_time_min']):
            df_master['actual_ride_time_min'] = pd.to_numeric(df_master['actual_ride_time_min'], errors='coerce')

            df_master['avg_speed_kmh'] = (
                df_master[distance_col] / (df_master['actual_ride_time_min'] / 60)
            ).replace([np.inf, -np.inf], np.nan)
            features_created.append('avg_speed_kmh')
            print(f"\n✅ avg_speed_kmh")
            print(f"   Mean speed: {df_master['avg_speed_kmh'].mean():.2f} km/h")
        else:
            print(f"\n⚠️ actual_ride_time_min is datetime, skipping speed calculation")

    # 6. Time Delay (Actual - Estimated)
    if 'actual_ride_time_min' in df_master.columns and 'estimated_ride_time_min' in df_master.columns:
        # Check both are not datetime
        if (not pd.api.types.is_datetime64_any_dtype(df_master['actual_ride_time_min']) and
            not pd.api.types.is_datetime64_any_dtype(df_master['estimated_ride_time_min'])):

            df_master['actual_ride_time_min'] = pd.to_numeric(df_master['actual_ride_time_min'], errors='coerce')
            df_master['estimated_ride_time_min'] = pd.to_numeric(df_master['estimated_ride_time_min'], errors='coerce')

            df_master['time_delay_min'] = (
                df_master['actual_ride_time_min'] - df_master['estimated_ride_time_min']
            )
            features_created.append('time_delay_min')
            print(f"\n✅ time_delay_min")
            print(f"   Mean delay: {df_master['time_delay_min'].mean():.2f} min")

            # Delay percentage
            df_master['time_delay_pct'] = (
                df_master['time_delay_min'] / df_master['estimated_ride_time_min'].replace(0, np.nan) * 100
            ).replace([np.inf, -np.inf], np.nan)
            features_created.append('time_delay_pct')
            print(f"✅ time_delay_pct")

            # Is Delayed
            df_master['is_delayed'] = (df_master['time_delay_min'] > 5).astype(int)
            features_created.append('is_delayed')
            delayed_pct = (df_master['is_delayed'].sum() / len(df_master)) * 100
            print(f"✅ is_delayed (>5 min late) - {delayed_pct:.1f}%")
        else:
            print(f"\n⚠️ Time columns are datetime format, skipping delay calculations")

    print(f"\n✅ Created {len(features_created)} distance-based features")

else:
    print("\n⚠️ No distance column found - skipping distance features")

print(f"\n📊 Current shape: {df_master.shape}")
print("="*70)
```

```python
"""
Cell 15: Create Location-Based Features
"""

print("📍 CREATING LOCATION-BASED FEATURES")
print("="*70)

features_created = []

# 1. City Pair (Pickup → Drop)
if 'city' in df_master.columns:
    # If single city column, it's the operating city
    df_master['city_cleaned'] = df_master['city'].astype(str).str.strip()
    features_created.append('city_cleaned')
    print(f"✅ city_cleaned")

    # City rank by popularity
    city_counts = df_master['city_cleaned'].value_counts()
    city_rank = {city: rank+1 for rank, city in enumerate(city_counts.index)}
    df_master['city_rank'] = df_master['city_cleaned'].map(city_rank)
    features_created.append('city_rank')
    print(f"✅ city_rank (1=most popular)")

# 2. Route (Pickup → Drop Location)
if 'pickup_location' in df_master.columns and 'drop_location' in df_master.columns:
    df_master['route'] = (
        df_master['pickup_location'].astype(str) + '_to_' +
        df_master['drop_location'].astype(str)
    )
    features_created.append('route')
    print(f"✅ route (pickup_to_drop)")

    # Is Same Location
    df_master['is_same_location'] = (
        df_master['pickup_location'] == df_master['drop_location']
    ).astype(int)
    features_created.append('is_same_location')
    same_loc_pct = (df_master['is_same_location'].sum() / len(df_master)) * 100
    print(f"✅ is_same_location ({same_loc_pct:.1f}% of rides)")

    # Popular route flag
    route_counts = df_master['route'].value_counts()
    popular_routes = route_counts.head(50).index  # Top 50 routes
    df_master['is_popular_route'] = df_master['route'].isin(popular_routes).astype(int)
    features_created.append('is_popular_route')
    print(f"✅ is_popular_route (top 50 routes)")

# 3. Demand Score (from location_demand if merged)
demand_cols = [col for col in df_master.columns if 'demand' in col.lower()]
if demand_cols:
    print(f"\n✅ Using demand columns: {demand_cols[:3]}...")

# 4. Traffic Level Encoding (if exists)
if 'traffic_level' in df_master.columns:
    traffic_mapping = {
        'Low': 1,
        'Medium': 2,
        'Moderate': 2,
        'High': 3,
        'Heavy': 3
    }
    df_master['traffic_level_encoded'] = df_master['traffic_level'].map(traffic_mapping)
    features_created.append('traffic_level_encoded')
    print(f"✅ traffic_level_encoded (1=Low, 3=High)")

# 5. Weather Encoding (if exists)
if 'weather_condition' in df_master.columns:
    weather_mapping = {
        'Clear': 1,
        'Sunny': 1,
        'Cloudy': 2,
        'Rainy': 3,
        'Rain': 3,
        'Heavy Rain': 4,
        'Storm': 4
    }
    df_master['weather_severity'] = df_master['weather_condition'].map(weather_mapping)
    features_created.append('weather_severity')
    print(f"✅ weather_severity (1=Clear, 4=Storm)")

    # Is Bad Weather
    df_master['is_bad_weather'] = (
        df_master['weather_condition'].str.lower().str.contains('rain|storm', na=False)
    ).astype(int)
    features_created.append('is_bad_weather')
    print(f"✅ is_bad_weather")

print(f"\n✅ Created {len(features_created)} location-based features")
print(f"📊 Current shape: {df_master.shape}")
print("="*70)
```

```python
"""
Cell 17: Create Customer Historical Features (Aggregated)
"""

print("👥 CREATING CUSTOMER AGGREGATED FEATURES")
print("="*70)

if 'customer_id' in df_master.columns:
    print("\nCalculating customer-level statistics...\n")

    features_created = []

    # 1. Customer Total Rides
    customer_ride_count = df_master.groupby('customer_id').size()
    df_master['cust_total_rides'] = df_master['customer_id'].map(customer_ride_count)
    features_created.append('cust_total_rides')
    print(f"✅ cust_total_rides")
    print(f"   Mean rides per customer: {df_master['cust_total_rides'].mean():.2f}")

    # 2. Customer Ride Frequency Category
    df_master['cust_frequency_category'] = pd.cut(
        df_master['cust_total_rides'],
        bins=[0, 1, 5, 10, 20, float('inf')],
        labels=['One-time', 'Occasional', 'Regular', 'Frequent', 'VIP']
    )
    features_created.append('cust_frequency_category')
    print(f"✅ cust_frequency_category")
    print(df_master['cust_frequency_category'].value_counts().sort_index())

    # 3. Customer Average Fare
    if fare_col:
        customer_avg_fare = df_master.groupby('customer_id')[fare_col].mean()
        df_master['cust_avg_fare'] = df_master['customer_id'].map(customer_avg_fare)
        features_created.append('cust_avg_fare')
        print(f"\n✅ cust_avg_fare")

    # 4. Customer Average Distance
    if distance_col:
        customer_avg_distance = df_master.groupby('customer_id')[distance_col].mean()
        df_master['cust_avg_distance'] = df_master['customer_id'].map(customer_avg_distance)
        features_created.append('cust_avg_distance')
        print(f"✅ cust_avg_distance")

    # 5. Customer Total Spend
    if fare_col:
        customer_total_spend = df_master.groupby('customer_id')[fare_col].sum()
        df_master['cust_total_spend'] = df_master['customer_id'].map(customer_total_spend)
        features_created.append('cust_total_spend')
        print(f"✅ cust_total_spend")

    # 6. Customer Cancellation Rate (if status available)
    if status_col:
        # Create cancellation flag
        df_master['is_cancelled'] = df_master[status_col].str.lower().str.contains('cancel', na=False).astype(int)

        customer_cancellations = df_master.groupby('customer_id')['is_cancelled'].sum()
        df_master['cust_total_cancellations'] = df_master['customer_id'].map(customer_cancellations)

        df_master['cust_cancellation_rate'] = (
            df_master['cust_total_cancellations'] / df_master['cust_total_rides'] * 100
        ).fillna(0)

        features_created.extend(['cust_total_cancellations', 'cust_cancellation_rate'])
        print(f"\n✅ cust_total_cancellations")
        print(f"✅ cust_cancellation_rate")
        print(f"   Mean cancellation rate: {df_master['cust_cancellation_rate'].mean():.2f}%")

        # High risk customer flag
        high_cancel_threshold = df_master['cust_cancellation_rate'].quantile(0.75)
        df_master['cust_is_high_cancel_risk'] = (
            df_master['cust_cancellation_rate'] >= high_cancel_threshold
        ).astype(int)
        features_created.append('cust_is_high_cancel_risk')
        print(f"✅ cust_is_high_cancel_risk (>={high_cancel_threshold:.1f}%)")

    # 7. Customer Loyalty Score (combination of rides and cancellation)
    if 'cust_total_rides' in df_master.columns and 'cust_cancellation_rate' in df_master.columns:
        # Normalize total rides (0-100)
        max_rides = df_master['cust_total_rides'].max()
        rides_score = (df_master['cust_total_rides'] / max_rides * 100).clip(upper=100)

        # Inverse of cancellation rate (0-100)
        cancel_penalty = (100 - df_master['cust_cancellation_rate']).clip(lower=0)

        # Weighted loyalty score
        df_master['cust_loyalty_score'] = (rides_score * 0.6 + cancel_penalty * 0.4).round(2)
        features_created.append('cust_loyalty_score')
        print(f"\n✅ cust_loyalty_score (0-100)")
        print(f"   Mean loyalty score: {df_master['cust_loyalty_score'].mean():.2f}")

    # 8. Customer Segment based on value
    if 'cust_total_spend' in df_master.columns:
        spend_25 = df_master['cust_total_spend'].quantile(0.25)
        spend_50 = df_master['cust_total_spend'].quantile(0.50)
        spend_75 = df_master['cust_total_spend'].quantile(0.75)

        df_master['cust_value_segment'] = pd.cut(
            df_master['cust_total_spend'],
            bins=[0, spend_25, spend_50, spend_75, float('inf')],
            labels=['Low Value', 'Medium Value', 'High Value', 'VIP']
        )
        features_created.append('cust_value_segment')
        print(f"✅ cust_value_segment")

    # 9. Is New Customer (only 1-2 rides)
    df_master['cust_is_new'] = (df_master['cust_total_rides'] <= 2).astype(int)
    features_created.append('cust_is_new')
    new_cust_pct = (df_master['cust_is_new'].sum() / len(df_master)) * 100
    print(f"\n✅ cust_is_new (<=2 rides) - {new_cust_pct:.1f}%")

    # 10. Customer Preferred Time (most common hour)
    if 'hour' in df_master.columns:
        customer_preferred_hour = df_master.groupby('customer_id')['hour'].agg(
            lambda x: x.mode()[0] if len(x.mode()) > 0 else -1
        )
        df_master['cust_preferred_hour'] = df_master['customer_id'].map(customer_preferred_hour)
        features_created.append('cust_preferred_hour')
        print(f"✅ cust_preferred_hour")

    print(f"\n✅ Created {len(features_created)} customer-level features")

else:
    print("⚠️ No customer_id column found")

print(f"\n📊 Current shape: {df_master.shape}")
print("="*70)
```

```python
"""
Cell 20: Create Driver Behavior Pattern Features
"""

print("🎯 CREATING DRIVER BEHAVIOR PATTERNS")
print("="*70)

if 'driver_id' in df_master.columns:

    features_created = []

    # 1. Driver Peak Hour Availability
    if 'is_rush_hour' in df_master.columns:
        driver_rush_rides = df_master[df_master['is_rush_hour'] == 1].groupby('driver_id').size()
        df_master['driver_rush_hour_rides'] = df_master['driver_id'].map(driver_rush_rides).fillna(0)

        df_master['driver_rush_hour_pct'] = (
            df_master['driver_rush_hour_rides'] / df_master['driver_total_rides'] * 100
        ).fillna(0)

        features_created.extend(['driver_rush_hour_rides', 'driver_rush_hour_pct'])
        print(f"✅ driver_rush_hour_rides, driver_rush_hour_pct")

    # 2. Driver Weekend Availability
    if 'is_weekend' in df_master.columns:
        driver_weekend_rides = df_master[df_master['is_weekend'] == 1].groupby('driver_id').size()
        df_master['driver_weekend_rides'] = df_master['driver_id'].map(driver_weekend_rides).fillna(0)

        df_master['driver_weekend_pct'] = (
            df_master['driver_weekend_rides'] / df_master['driver_total_rides'] * 100
        ).fillna(0)

        features_created.extend(['driver_weekend_rides', 'driver_weekend_pct'])
        print(f"✅ driver_weekend_rides, driver_weekend_pct")

    # 3. Driver Late Night Rides
    if 'is_late_night' in df_master.columns:
        driver_late_night_rides = df_master[df_master['is_late_night'] == 1].groupby('driver_id').size()
        df_master['driver_late_night_rides'] = df_master['driver_id'].map(driver_late_night_rides).fillna(0)

        df_master['driver_late_night_pct'] = (
            df_master['driver_late_night_rides'] / df_master['driver_total_rides'] * 100
        ).fillna(0)

        features_created.extend(['driver_late_night_rides', 'driver_late_night_pct'])
        print(f"✅ driver_late_night_rides, driver_late_night_pct")

    # 4. Driver Long Distance Preference
    if 'is_long_ride' in df_master.columns:
        driver_long_rides = df_master[df_master['is_long_ride'] == 1].groupby('driver_id').size()
        df_master['driver_long_rides'] = df_master['driver_id'].map(driver_long_rides).fillna(0)

        df_master['driver_long_ride_pct'] = (
            df_master['driver_long_rides'] / df_master['driver_total_rides'] * 100
        ).fillna(0)

        features_created.extend(['driver_long_rides', 'driver_long_ride_pct'])
        print(f"✅ driver_long_rides, driver_long_ride_pct")

    # 5. Driver Average Speed (if speed calculated)
    if 'avg_speed_kmh' in df_master.columns:
        driver_avg_speed = df_master.groupby('driver_id')['avg_speed_kmh'].mean()
        df_master['driver_typical_speed'] = df_master['driver_id'].map(driver_avg_speed)
        features_created.append('driver_typical_speed')
        print(f"\n✅ driver_typical_speed")

    # 6. Driver Delay Rate (if delay calculated)
    if 'is_delayed' in df_master.columns:
        driver_delays = df_master[df_master['is_delayed'] == 1].groupby('driver_id').size()
        df_master['driver_total_delays'] = df_master['driver_id'].map(driver_delays).fillna(0)

        df_master['driver_delay_rate'] = (
            df_master['driver_total_delays'] / df_master['driver_total_rides'] * 100
        ).fillna(0)

        features_created.extend(['driver_total_delays', 'driver_delay_rate'])
        print(f"✅ driver_total_delays, driver_delay_rate")

    # 7. Driver Revenue per Ride
    if 'driver_total_revenue' in df_master.columns and 'driver_total_rides' in df_master.columns:
        df_master['driver_revenue_per_ride'] = (
            df_master['driver_total_revenue'] / df_master['driver_total_rides']
        ).round(2)
        features_created.append('driver_revenue_per_ride')
        print(f"✅ driver_revenue_per_ride")

    # 8. Driver Primary Operating City
    if 'city' in df_master.columns:
        driver_primary_city = df_master.groupby('driver_id')['city'].agg(
            lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown'
        )
        df_master['driver_primary_city'] = df_master['driver_id'].map(driver_primary_city)
        features_created.append('driver_primary_city')
        print(f"✅ driver_primary_city")

    # 9. Driver Utilization Score (rides per day active)
    # This is a proxy - actual implementation would need date ranges
    if 'driver_total_rides' in df_master.columns:
        # Simple score based on total rides (proxy for activity)
        max_rides = df_master['driver_total_rides'].max()
        df_master['driver_utilization_score'] = (
            (df_master['driver_total_rides'] / max_rides) * 100
        ).round(2).clip(upper=100)
        features_created.append('driver_utilization_score')
        print(f"\n✅ driver_utilization_score (0-100)")

    print(f"\n✅ Created {len(features_created)} driver behavior features")

else:
    print("⚠️ No driver_id column found")

print(f"\n📊 Current shape: {df_master.shape}")
print("="*70)
```

```python
"""
Cell 21: Create Customer-Driver Interaction Features
"""

print("🤝 CREATING CUSTOMER-DRIVER INTERACTION FEATURES")
print("="*70)

if 'customer_id' in df_master.columns and 'driver_id' in df_master.columns:
    print("\nAnalyzing customer-driver interactions...\n")

    features_created = []

    # 1. Customer-Driver Match History (have they ridden together before?)
    customer_driver_pairs = df_master.groupby(['customer_id', 'driver_id']).size()
    df_master['cust_driver_pair_id'] = (
        df_master['customer_id'].astype(str) + '_' + df_master['driver_id'].astype(str)
    )

    pair_counts = df_master['cust_driver_pair_id'].map(
        df_master['cust_driver_pair_id'].value_counts()
    )
    df_master['cust_driver_ride_count'] = pair_counts
    features_created.append('cust_driver_ride_count')
    print(f"✅ cust_driver_ride_count (how many times together)")

    # 2. Is Repeat Pair
    df_master['is_repeat_pair'] = (df_master['cust_driver_ride_count'] > 1).astype(int)
    features_created.append('is_repeat_pair')
    repeat_pct = (df_master['is_repeat_pair'].sum() / len(df_master)) * 100
    print(f"✅ is_repeat_pair - {repeat_pct:.1f}% are repeat pairings")

    # 3. Customer-Driver Experience Match
    if 'cust_total_rides' in df_master.columns and 'driver_total_rides' in df_master.columns:
        # Both new, both experienced, or mismatched
        df_master['cust_driver_exp_match'] = 'Mismatched'

        # Both new
        df_master.loc[
            (df_master['cust_total_rides'] <= 5) & (df_master['driver_total_rides'] <= 20),
            'cust_driver_exp_match'
        ] = 'Both New'

        # Both experienced
        df_master.loc[
            (df_master['cust_total_rides'] > 10) & (df_master['driver_total_rides'] > 100),
            'cust_driver_exp_match'
        ] = 'Both Experienced'

        features_created.append('cust_driver_exp_match')
        print(f"✅ cust_driver_exp_match")
        print(df_master['cust_driver_exp_match'].value_counts())

    # 4. Risk Score (High cancel customer + Unreliable driver)
    if 'cust_is_high_cancel_risk' in df_master.columns and 'driver_is_unreliable' in df_master.columns:
        df_master['high_risk_pairing'] = (
            (df_master['cust_is_high_cancel_risk'] == 1) &
            (df_master['driver_is_unreliable'] == 1)
        ).astype(int)
        features_created.append('high_risk_pairing')
        high_risk_pct = (df_master['high_risk_pairing'].sum() / len(df_master)) * 100
        print(f"\n✅ high_risk_pairing - {high_risk_pct:.1f}% high risk")

    # 5. Premium Pairing (VIP customer + Top driver)
    if 'cust_frequency_category' in df_master.columns and 'driver_performance_tier' in df_master.columns:
        df_master['is_premium_pairing'] = (
            (df_master['cust_frequency_category'] == 'VIP') &
            (df_master['driver_performance_tier'] == 'Platinum')
        ).astype(int)
        features_created.append('is_premium_pairing')
        premium_pct = (df_master['is_premium_pairing'].sum() / len(df_master)) * 100
        print(f"✅ is_premium_pairing - {premium_pct:.1f}% premium")

    # 6. Customer Loyalty vs Driver Reliability Score
    if 'cust_loyalty_score' in df_master.columns and 'driver_reliability_score' in df_master.columns:
        df_master['pairing_quality_score'] = (
            (df_master['cust_loyalty_score'] + df_master['driver_reliability_score']) / 2
        ).round(2)
        features_created.append('pairing_quality_score')
        print(f"✅ pairing_quality_score (0-100)")
        print(f"   Mean: {df_master['pairing_quality_score'].mean():.2f}")

    # 7. Price Sensitivity vs Fare Match
    if 'cust_price_sensitivity' in df_master.columns and fare_col:
        # Is customer price sensitive but got high fare?
        overall_avg = df_master[fare_col].mean()

        df_master['price_mismatch'] = (
            (df_master['cust_price_sensitivity'] < -10) &  # Price sensitive
            (df_master[fare_col] > overall_avg * 1.2)  # High fare
        ).astype(int)

        features_created.append('price_mismatch')
        mismatch_pct = (df_master['price_mismatch'].sum() / len(df_master)) * 100
        print(f"\n✅ price_mismatch - {mismatch_pct:.1f}% mismatches")

    # 8. Customer Preferred Time vs Actual Time Match
    if 'cust_preferred_hour' in df_master.columns and 'hour' in df_master.columns:
        df_master['time_preference_match'] = (
            df_master['cust_preferred_hour'] == df_master['hour']
        ).astype(int)
        features_created.append('time_preference_match')
        match_pct = (df_master['time_preference_match'].sum() / len(df_master)) * 100
        print(f"✅ time_preference_match - {match_pct:.1f}% matches")

    print(f"\n✅ Created {len(features_created)} interaction features")

else:
    print("⚠️ Customer or Driver ID not found")

print(f"\n📊 Current shape: {df_master.shape}")
print("="*70)
```

```python
"""
Cell 22: Create Context-Based Interaction Features
"""

print("🌟 CREATING CONTEXT-BASED INTERACTION FEATURES")
print("="*70)

features_created = []

# 1. Demand-Supply Mismatch
if 'is_rush_hour' in df_master.columns:
    # During rush hour with new driver = potential issue
    if 'driver_is_new' in df_master.columns:
        df_master['rush_hour_new_driver'] = (
            (df_master['is_rush_hour'] == 1) & (df_master['driver_is_new'] == 1)
        ).astype(int)
        features_created.append('rush_hour_new_driver')
        print(f"✅ rush_hour_new_driver")

    # Rush hour with new customer = education needed
    if 'cust_is_new' in df_master.columns:
        df_master['rush_hour_new_customer'] = (
            (df_master['is_rush_hour'] == 1) & (df_master['cust_is_new'] == 1)
        ).astype(int)
        features_created.append('rush_hour_new_customer')
        print(f"✅ rush_hour_new_customer")

# 2. Weather-Experience Interaction
if 'is_bad_weather' in df_master.columns and 'driver_experience_level' in df_master.columns:
    df_master['bad_weather_newbie_driver'] = (
        (df_master['is_bad_weather'] == 1) &
        (df_master['driver_experience_level'].isin(['Newbie', 'Beginner']))
    ).astype(int)
    features_created.append('bad_weather_newbie_driver')
    print(f"✅ bad_weather_newbie_driver")

# 3. Long Ride + Late Night
if 'is_long_ride' in df_master.columns and 'is_late_night' in df_master.columns:
    df_master['long_late_night_ride'] = (
        (df_master['is_long_ride'] == 1) & (df_master['is_late_night'] == 1)
    ).astype(int)
    features_created.append('long_late_night_ride')
    late_long_pct = (df_master['long_late_night_ride'].sum() / len(df_master)) * 100
    print(f"✅ long_late_night_ride - {late_long_pct:.1f}%")

# 4. Surge + Price Sensitive Customer
if 'has_surge' in df_master.columns and 'cust_price_sensitivity' in df_master.columns:
    df_master['surge_price_sensitive_cust'] = (
        (df_master['has_surge'] == 1) & (df_master['cust_price_sensitivity'] < -10)
    ).astype(int)
    features_created.append('surge_price_sensitive_cust')
    print(f"✅ surge_price_sensitive_cust")

# 5. High Traffic + Inexperienced Driver
if 'traffic_level_encoded' in df_master.columns and 'driver_experience_level' in df_master.columns:
    df_master['high_traffic_new_driver'] = (
        (df_master['traffic_level_encoded'] >= 3) &
        (df_master['driver_experience_level'].isin(['Newbie', 'Beginner']))
    ).astype(int)
    features_created.append('high_traffic_new_driver')
    print(f"✅ high_traffic_new_driver")

# 6. Weekend + Surge
if 'is_weekend' in df_master.columns and 'has_surge' in df_master.columns:
    df_master['weekend_surge'] = (
        (df_master['is_weekend'] == 1) & (df_master['has_surge'] == 1)
    ).astype(int)
    features_created.append('weekend_surge')
    print(f"✅ weekend_surge")

# 7. Business Hours + High Fare
if 'is_business_hours' in df_master.columns and 'is_high_fare' in df_master.columns:
    df_master['business_hours_premium'] = (
        (df_master['is_business_hours'] == 1) & (df_master['is_high_fare'] == 1)
    ).astype(int)
    features_created.append('business_hours_premium')
    print(f"✅ business_hours_premium")

# 8. Same Location + Short Distance
if 'is_same_location' in df_master.columns and 'is_short_ride' in df_master.columns:
    df_master['same_loc_short_ride'] = (
        (df_master['is_same_location'] == 1) & (df_master['is_short_ride'] == 1)
    ).astype(int)
    features_created.append('same_loc_short_ride')
    print(f"✅ same_loc_short_ride")

# 9. Popular Route + Rush Hour
if 'is_popular_route' in df_master.columns and 'is_rush_hour' in df_master.columns:
    df_master['popular_route_rush_hour'] = (
        (df_master['is_popular_route'] == 1) & (df_master['is_rush_hour'] == 1)
    ).astype(int)
    features_created.append('popular_route_rush_hour')
    print(f"✅ popular_route_rush_hour")

# 10. Combined Risk Score
risk_factors = []
if 'high_risk_pairing' in df_master.columns:
    risk_factors.append('high_risk_pairing')
if 'bad_weather_newbie_driver' in df_master.columns:
    risk_factors.append('bad_weather_newbie_driver')
if 'high_traffic_new_driver' in df_master.columns:
    risk_factors.append('high_traffic_new_driver')
if 'surge_price_sensitive_cust' in df_master.columns:
    risk_factors.append('surge_price_sensitive_cust')

if risk_factors:
    df_master['total_risk_factors'] = df_master[risk_factors].sum(axis=1)
    features_created.append('total_risk_factors')
    print(f"\n✅ total_risk_factors (0-{len(risk_factors)})")
    print(f"   Mean: {df_master['total_risk_factors'].mean():.2f}")

print(f"\n✅ Created {len(features_created)} context-based features")
print(f"📊 Current shape: {df_master.shape}")
print("="*70)
```

```python
"""
Cell 23: Create Target Variables for Modeling
"""

print("🎯 CREATING TARGET VARIABLES")
print("="*70)

if status_col:
    print(f"\nUsing status column: {status_col}\n")

    # 1. Multi-class Target: Booking Outcome
    print("1️⃣ MULTI-CLASS TARGET: Booking Outcome")
    print("-" * 70)

    # Clean status values
    df_master['booking_outcome'] = df_master[status_col].str.strip().str.title()

    # Show distribution
    print("Original status distribution:")
    print(df_master['booking_outcome'].value_counts())

    # Map to standard categories
    outcome_mapping = {}
    for status in df_master['booking_outcome'].unique():
        status_lower = str(status).lower()
        if 'complet' in status_lower:
            outcome_mapping[status] = 'Completed'
        elif 'cancel' in status_lower:
            outcome_mapping[status] = 'Cancelled'
        elif 'incomplete' in status_lower or 'fail' in status_lower:
            outcome_mapping[status] = 'Incomplete'
        else:
            outcome_mapping[status] = 'Other'

    df_master['target_outcome'] = df_master['booking_outcome'].map(outcome_mapping)

    print("\n✅ target_outcome created:")
    print(df_master['target_outcome'].value_counts())
    print(f"\nClass distribution:")
    for outcome, count in df_master['target_outcome'].value_counts().items():
        pct = (count / len(df_master)) * 100
        print(f"   {outcome}: {count:,} ({pct:.2f}%)")

    # Encode for modeling
    from sklearn.preprocessing import LabelEncoder
    le_outcome = LabelEncoder()
    df_master['target_outcome_encoded'] = le_outcome.fit_transform(df_master['target_outcome'])

    print(f"\n✅ target_outcome_encoded created (numeric)")
    print(f"   Mapping: {dict(enumerate(le_outcome.classes_))}")

    # 2. Binary Target: Is Cancelled
    print("\n" + "=" * 70)
    print("2️⃣ BINARY TARGET: Is Cancelled")
    print("-" * 70)

    df_master['target_is_cancelled'] = (
        df_master['target_outcome'] == 'Cancelled'
    ).astype(int)

    cancelled_count = df_master['target_is_cancelled'].sum()
    cancelled_pct = (cancelled_count / len(df_master)) * 100

    print(f"✅ target_is_cancelled created:")
    print(f"   Cancelled: {cancelled_count:,} ({cancelled_pct:.2f}%)")
    print(f"   Not Cancelled: {(len(df_master) - cancelled_count):,} ({(100-cancelled_pct):.2f}%)")

    # 3. Binary Target: Customer Cancelled (if available)
    # Check if we can identify who cancelled
    cancel_by_col = None
    for col in df_master.columns:
        if 'cancelled_by' in col.lower() or 'canceled_by' in col.lower():
            cancel_by_col = col
            break

    if cancel_by_col:
        print("\n" + "=" * 70)
        print("3️⃣ BINARY TARGET: Customer Cancelled")
        print("-" * 70)
```

```python
"""
Cell 24: Analyze Class Imbalance in Target Variables
"""

print("⚖️ ANALYZING CLASS IMBALANCE")
print("="*70)

imbalance_report = []

# Check each target variable
target_vars = [
    ('target_outcome', 'multi-class'),
    ('target_is_cancelled', 'binary'),
    ('target_customer_cancelled', 'binary'),
    ('target_driver_delayed', 'binary')
]

for target_name, target_type in target_vars:
    if target_name in df_master.columns:
        print(f"\n📊 {target_name} ({target_type}):")
        print("-" * 70)

        value_counts = df_master[target_name].value_counts()

        if target_type == 'binary':
            minority_count = value_counts.min()
            majority_count = value_counts.max()
            imbalance_ratio = majority_count / minority_count

            print(f"   Majority class: {majority_count:,} samples")
            print(f"   Minority class: {minority_count:,} samples")
            print(f"   Imbalance ratio: {imbalance_ratio:.2f}:1")

            if imbalance_ratio > 10:
                print(f"   ⚠️ SEVERE IMBALANCE - Consider SMOTE or class weights")
            elif imbalance_ratio > 3:
                print(f"   ⚠️ MODERATE IMBALANCE - Consider class weights")
            else:
                print(f"   ✅ BALANCED - No special handling needed")

            imbalance_report.append({
                'Target': target_name,
                'Type': target_type,
                'Imbalance_Ratio': imbalance_ratio,
                'Recommendation': 'SMOTE' if imbalance_ratio > 10 else 'Class Weights' if imbalance_ratio > 3 else 'None'
            })

        else:  # multi-class
            print("   Class distribution:")
            for cls, count in value_counts.items():
                pct = (count / len(df_master)) * 100
                print(f"      {cls}: {count:,} ({pct:.2f}%)")

            # Check if any class is very small
            min_class_pct = (value_counts.min() / len(df_master)) * 100

            if min_class_pct < 5:
                print(f"   ⚠️ Smallest class is only {min_class_pct:.2f}% - Consider combining classes")
            else:
                print(f"   ✅ All classes have reasonable representation")

# Summary table
if imbalance_report:
    print("\n" + "=" * 70)
    print("📋 IMBALANCE SUMMARY:")
    print("=" * 70)

    import pandas as pd
    imbalance_df = pd.DataFrame(imbalance_report)
    display(imbalance_df)

    print("\n💡 RECOMMENDATIONS:")
    for idx, row in imbalance_df.iterrows():
        if row['Recommendation'] != 'None':
            print(f"   • {row['Target']}: Use {row['Recommendation']}")

print("\n" + "=" * 70)
```
